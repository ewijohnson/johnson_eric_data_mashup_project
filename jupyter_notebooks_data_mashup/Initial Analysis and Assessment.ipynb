```python
import pandas as pd
import numpy as np
import bqplot
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
final_dataset_file = '../Final Dataset/johnson_eric_final_dataset.csv'
final_dataset = pd.read_csv(final_dataset_file, index_col=0)
```

```python
final_dataset.head(20)
```

```python
game_soundtrack_ratings = final_dataset.loc[:, ['Game', 'Critic_Score', 'mp3 Downloads Rank']]
game_soundtrack_ratings = game_soundtrack_ratings.groupby([game_soundtrack_ratings['Game']]).mean()
game_soundtrack_ratings.head(20)
```

```python
game_soundtrack_ratings = game_soundtrack_ratings[pd.notnull(game_soundtrack_ratings['Critic_Score'])]
game_soundtrack_ratings = game_soundtrack_ratings[pd.notnull(game_soundtrack_ratings['mp3 Downloads Rank'])]
game_soundtrack_ratings.head(20)
```

The y axis was reversed so as to show any correlation going from lower left corner to upper right corner$.^{1}$

```python
x = game_soundtrack_ratings['Critic_Score']
y = game_soundtrack_ratings['mp3 Downloads Rank']

plt.scatter(x, y)
plt.gca().invert_yaxis()
plt.title('Critic Score vs Downloads Rank')
plt.xlabel('Critic Score')
plt.ylabel('mp3 Downloads Rank')
plt.show()
```

```python
game_soundtrack_user = final_dataset.loc[:, ['Game', 'User_Score', 'mp3 Downloads Rank']]
game_soundtrack_user = game_soundtrack_user[game_soundtrack_user['User_Score'] != 'tbd']

game_soundtrack_user['User_Score'] = game_soundtrack_user['User_Score'].astype('float64')

game_soundtrack_user = game_soundtrack_user[pd.notnull(game_soundtrack_user['User_Score'])]
game_soundtrack_user = game_soundtrack_user[pd.notnull(game_soundtrack_user['mp3 Downloads Rank'])]


game_soundtrack_user = game_soundtrack_user.groupby([game_soundtrack_user['Game']]).mean()

game_soundtrack_user.head(20)
```

```python
x = game_soundtrack_user['User_Score']
y = game_soundtrack_user['mp3 Downloads Rank']

plt.scatter(x, y)
plt.gca().invert_yaxis()
plt.title('User Score vs Downloads Rank')
plt.xlabel('User Score')
plt.ylabel('mp3 Downloads Rank')
plt.show()
```

```python
genre_popularity = final_dataset.loc[:, ['Genre', 'VGMDB Soundtrack Popularity']]
genre_popularity = genre_popularity[pd.notnull(genre_popularity['VGMDB Soundtrack Popularity'])]
genre_popularity = genre_popularity.groupby([genre_popularity['Genre']]).mean()
genre_popularity
```

```python
x = genre_popularity.index
y = genre_popularity['VGMDB Soundtrack Popularity']

plt.bar(x, y)
plt.title('Popularity by Genre')
plt.xticks(rotation=-45)
plt.xlabel('Genre')
plt.ylabel('Soundtrack Popularity')
plt.show()
```

```python
composer_genre = final_dataset.loc[:, ['Composer', 'Genre', 'Game']]
composer_genre = composer_genre.groupby([composer_genre['Composer'], composer_genre['Genre']]).count()
composer_genre = composer_genre.reset_index(level=[0,1])
composer_genre = composer_genre.rename(columns={'Game': 'Game Count'})
composer_genre.head(20)
```

```python
len(composer_genre.index)
```

```python
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
c_sc = bqplot.ColorScale()

x_ax = bqplot.Axis(scale=x_sc, tick_rotate=-45, label='Composer')
y_ax = bqplot.Axis(scale=y_sc, label='Genre', orientation='vertical')
c_ax = bqplot.ColorAxis(scale=c_sc, label='Number(Log)', orientation='vertical', side='right')

scatters = bqplot.Scatter(x=composer_genre['Composer'], y=composer_genre['Genre'], 
                         color=np.log10(composer_genre['Game Count']),
                         scales={'x': x_sc, 'y': y_sc, 'color': c_sc})

fig = bqplot.Figure(marks=[scatters], axes=[x_ax, y_ax, c_ax], 
                    title='Number of Games Each Composer Worked on by Genre (All 201)')
fig
```

```python
composer_genre1 = composer_genre[:102]

x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
c_sc = bqplot.ColorScale()

x_ax = bqplot.Axis(scale=x_sc, tick_rotate=-45, label='Composer')
y_ax = bqplot.Axis(scale=y_sc, label='Genre', orientation='vertical')
c_ax = bqplot.ColorAxis(scale=c_sc, label='Number(Log)', orientation='vertical', side='right')

scatters = bqplot.Scatter(x=composer_genre1['Composer'], y=composer_genre1['Genre'], 
                         color=np.log10(composer_genre1['Game Count']),
                         scales={'x': x_sc, 'y': y_sc, 'color': c_sc})

fig = bqplot.Figure(marks=[scatters], axes=[x_ax, y_ax, c_ax], 
                    title='Number of Games Each Composer Worked on by Genre (1-101)')
fig
```
